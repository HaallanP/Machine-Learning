ESTUDO SOBRE MACHINE LEANING


  * **KNeighborsClassifier**
    * Vizinhança: O algoritmo classifica um novo exemplo com base nos rótulos dos seus vizinhos mais próximos no espaço de características. O "K" se refere ao número de vizinhos que o algoritmo considera para fazer a previsão.

  * **MinMaxScaler**
    * O MinMaxScaler é uma técnica de normalização de dados utilizada para escalar características para um intervalo específico, geralmente entre 0 e 1. Isso é especialmente útil em algoritmos de aprendizado de máquina que são sensíveis à escala dos dados, como o KNeighborsClassifier.

```
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Estudos /Machine Learning/exemplo2.csv')
df.head()
```

```
#Separando a coluna risco
X = df.drop('risco', axis=1)
y = df.risco
```

```
#treinando as colunas
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X,y)#Treinando
```

```
#Prevendo
predictions = knn.predict([[18,1000]])
display(f'{predictions}')
```

```
#Treinando os dados
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.30, random_state=42)
```

```
#Treinado
knn2 = KNeighborsClassifier(n_neighbors=5,
                            leaf_size=30,
                            weights='uniform')
knn2.fit(X_train, y_train)#Treino
knn2 = knn2.predict(X_test)#Prevendo
```

```
#Acuracia
acuracia = accuracy_score(y_test, knn2) * 100  #1°- O valor do teste y_test, 2° - O valor da predição knn2
print(f'{acuracia:.2f}%')
```

## Colocanco peso sobre os dados

#####Usaremos **MinMaxScaler** chamado **Normalizador**, ultlizaremos **MinMaxScaler.fit_transform()** para ajustados o normalizador.

```
from sklearn.preprocessing import MinMaxScaler
```

```
#Classificador MinMazScaler
normalizador = MinMaxScaler()
X_norma = normalizador.fit_transform(X)#usando a varivavel do inicio demos o valor de X
X_norma
```

```
#Treinando os novos dados
X_train, X_test, y_train, y_test = train_test_split(X_norma, y, train_size=0.30, random_state=42)
knn3 = KNeighborsClassifier(n_neighbors=7,
                            leaf_size=30,
                            weights='uniform')

knn3.fit(X_train, y_train)#Treino
knn33 = knn3.predict(X_test)#Prevendo
```

```
#Acuracia
acuracia = accuracy_score(y_test, knn33) * 100 #1°- O valor do teste y_test, 2° - O valor da predição knn3
print(f'{acuracia:.2f}%')
```

```
#Previsão com dados normalizados

novo_cliente = [[18,1000]]
X_new = normalizador.transform(novo_cliente)
res = knn3.predict(X_new)
display(f'{res}')
```
